```python
from pyproj import Proj, transform, Transformer
import pandas as pd
import numpy as np
import os
```

```python
etrs89_utm = "epsg:3763"  # ETRS89 UTM zona 29N
wgs84_utm = "epsg:32629"   # WGS84 UTM zona 29N
wgs84_deg = "epsg:4326" # WGS84 UTM zona 29N

transformer = Transformer.from_crs(etrs89_utm, wgs84_utm, always_xy=True)

# Converter GMS para Graus Decimais
def gms_para_decimal(graus, minutos, segundos, direcao):
    # Calcula o valor decimal positivo
    decimal = graus + (minutos / 60.0) + (segundos / 3600.0)
    
    # Se for Sul ou Oeste, o valor fica negativo
    if direcao.upper() in ['S', 'W', 'O']:
        decimal = -decimal
        
    return decimal

# coordenadas GMS
lat, lon = gms_para_decimal(41, 30, 0, 'N'), gms_para_decimal(8, 52, 50, 'W')
print(f"Latitude em decimal: {lat}, Longitude em decimal: {lon}")
# coordenadas ETRS89
etrsX, etrsY =  -60000 , 195000

# Converter para WGS84 UTM
x_end, y_end = transformer.transform(etrsX, etrsY)
# x_end, y_end = transformer.transform(lon, lat)



x_end, y_end
```

# Arquivo xy

```python
file_path = r'D:\XBeach\Esposende\new\spectral input\area analise\sanstrip_south_groines.txt'
lat_long = pd.read_csv(file_path, skiprows=0,sep=r'\s+', names=['long', 'lat'])
lat_long
```

```python
etrs89_utm = "epsg:3763"  # ETRS89 UTM zona 29N
wgs84_utm = "epsg:32629"   # WGS84 UTM zona 29N
wgs84_deg = "epsg:4326" # WGS84 UTM zona 29N

transformer = Transformer.from_crs(wgs84_deg, wgs84_utm, always_xy=True)

# Separar coordenadas
x_orig, y_orig = lat_long.iloc[:, 0], lat_long.iloc[:, 1]

# Converter para WGS84 UTM
x_end, y_end = transformer.transform(x_orig, y_orig)
xy_end = np.column_stack((x_end, y_end))

# # Salvar o novo arquivo convertido
output_dir = os.path.dirname(file_path)
base_name = os.path.splitext(os.path.basename(file_path))[0]
np.savetxt( os.path.join(output_dir, f'{base_name}_UTM.xy'), xy_end, fmt="%.6f", delimiter="\t")
print(f"Conversão concluída! Arquivo salvo em {output_dir} como {base_name}_UTM.xy.")
```
